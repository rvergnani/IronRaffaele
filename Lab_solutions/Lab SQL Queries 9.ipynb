```python
films_rented_may = pd.read_sql("SELECT customer_id, COUNT(rental_id) AS films_rented_may FROM rentals_may GROUP BY customer_id", engine)
```

```python
films_rented_june = pd.read_sql("SELECT customer_id, COUNT(rental_id) AS films_rented_june FROM rentals_june GROUP BY customer_id", engine)
```

```python
help(pd.merge)
```

```python
file = pd.merge(films_rented_may, films_rented_june, on = "customer_id")
```

```python
file
```

```python
import numpy as np
```

```python
conditions = [
    (file["films_rented_may"] > file["films_rented_june"]),
    (file["films_rented_may"] < file["films_rented_june"]),
    (file["films_rented_may"] == file["films_rented_june"])]
```

```python
values = ["May", "June", "Same in May and June"]
```

```python
file["Month with most rentals"] = np.select(conditions, values)
```
